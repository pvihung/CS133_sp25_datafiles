# CL4.1-In class exercise


```python
import pandas as pd
csvFile = 'https://raw.githubusercontent.com/csbfx/advpy122-data/master/top_movies_2020.csv'
movies = pd.read_csv(csvFile)
movies
```

```python
movies.Year
''' Dau [] dung de lay data trong dataframe '''
movies_between_1930_and_1970s = movies[(movies['Year'] > 1930) & (movies['Year'] < 1980)]
print(movies_between_1930_and_1970s.describe())
```

```python
movies.Gross
''' idxmax() dung de lay index lon nhat trong dataframe '''
movies_with_highest_gross = movies.loc[movies['Gross'].idxmax(), 'Title']
print(movies_with_highest_gross)
```

# Introduction to Pandas


### Create data
Có 2 core objets trong Pandas:
- **Dataframe**: Nói chung là giống spreadsheet và database table, chứa cột và cột có thể duplicate
- **Series**: là kiểu dữ liệu 1 chiều (cột) chứa một dãy giá tri, tưởng tượng nó giống table khi bấm describe()

```python
mylist = ['a', 'b', 'c']
pd.Series(mylist)
```

```python
mydataset = {
    'cars': ["BMW", "Volvo", "Ford"],
    'passings': [3, 7, 2]
}
pd.DataFrame(mydataset)
```

## Có 2 kiểu khi tạo dataframe
- Dùng []: khi dùng [] thì sẽ thêm các variable theo chiều ngang (row)

- Dùng {}: khi dùng {} thì sẽ thêm các variable theo chiều dọc (column), nhung khi thi phai tu assign cai nao thuoc column nao

```python
#VD1
list1 = ['a', 'b', 'c']
list2 = [1, 2, 3]
pd.DataFrame([list1, list2])
```

```python
#VD2
lista = ['a', 'b', 'c']
listb = [1, 2, 3]
pd.DataFrame({'A': lista, 'B': listb})
```

Ngoai ra, ta co the tu assignment Index cua DataFrame

```python
productList = ['Product A', 'Product B', 'Product C']
pd.DataFrame({'Names': list1, 'Scores' : list2}, index = productList)
```
